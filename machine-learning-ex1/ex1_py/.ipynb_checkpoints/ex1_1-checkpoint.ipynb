```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('ex1data1.txt', delimiter=',')
```

```python
X = data[:, 0]
y = data[:, 1]
```

```python
m = len(X)
```

```python
data = np.insert(data, 0, 1, 1)
X = data[:,0:2]

theta = np.zeros((2, 1))
iterations = 1500
alpha = 0.01
```

```python
y = y.reshape(97, 1)
```

```python
def computecost(X, y, theta, m):
    J_theta=np.sum(np.power((np.dot(X,theta)-y), 2))/(2*m)
    return J_theta
```

```python
def gradientdescent(X, y, theta, m, iterations, alpha):
    temp = np.matrix(np.zeros(theta.shape))
    cost = np.zeros(iterations)
    for i in range(iterations):
        error = np.dot(X,theta)-y
        for j in range(theta.shape[0]):
            temp[j,0] = theta[j,0] - alpha*np.sum(np.multiply(error,X[:,j].reshape(97, 1)))/m
        theta = temp
        cost[i] = computecost(X, y, theta ,m)
    return theta, cost
```

```python
theta1, cost= gradientdescent(X, y, theta, m, iterations, alpha)
theta1, cost
```

```python
predict1 = [1, 3.5] * theta1
predict2 = [1, 7] * theta1
print(predict1)
print(predict2)
```

```python
x = np.linspace(X[:,1].min(), X.max(), 100)
f = theta1[0, 0] + (theta1[1, 0] * x)

fig, ax = plt.subplots(figsize=(12,8))
ax.plot(x, f, 'r', label='Prediction')
ax.plot(X[:,1], y, 'x', label='Traning Data')
ax.legend(loc=2)
ax.set_ylabel('Profit in $10,000s')
ax.set_xlabel('Population of City in 10,000s')
ax.set_title('Training data with linear regression')
plt.show()
```
